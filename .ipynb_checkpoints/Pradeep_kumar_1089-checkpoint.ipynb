```python
y=df['diagnosis']
```

```python
y
```

```python
X=df.drop(columns=['id','diagnosis'])
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
K=[]
```

```python
training=[]
```

```python
test=[]
```

```python
scores={}
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for k in range(2,21):
    clf=KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train,y_train)
    training_score=clf.score(X_train,y_train)
    test_score=clf.score(X_test,y_test)
    K.append(k)
    training.append(training_score)
    test.append(test_score)
    
scores[k]=[training_score,test_score]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(K,training)
plt.savefig('variationOfK.png')
plt.show()
```

```python
plt.scatter(K,training)
plt.scatter(K,test)
plt.savefig('valueOf_K.png')
plt.show()
```

```python
clf=KNeighborsClassifier(n_neighbors=7)
clf.fit(X_train,y_train)
y.pred=clf.predict(X_test)
```

```python
y.pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y.pred,y_test))
```

```python
print(confusion_matrix(y.pred,y_test))
```

```python
print(classification_report(y.pred,y_test))
```
